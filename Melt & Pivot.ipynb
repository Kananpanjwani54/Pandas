```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92],
    'Science': [90, 82, 89],
    'English': [88, 85, 94]
}

df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
#represent same info in long format ,it uses melt method
#basically melt() takes the columns that represent different variables and combines them into key value pair.

df1=df.melt(id_vars=["Name"], value_vars=["Math", "Science", "English"], var_name="Subject", value_name="Score")
```

```python
df1
```

```python
df1.pivot(index="Name", columns="Subject", values="Score")
```

```python
# wide is used to convert long format data into wide format data 
df1.pivot(index="Name", columns="Subject", values="Score").reset_index()
```

```python
df1.pivot(index="Name", columns="Subject", values="Score").rename_axis(columns=None)
```

# Pivot Function 
## Using pivot() to reshape it into wide format.
## If you have duplicate values in tale.
## We can use pivot_table() to aggregate values (e.g., taking the mean for duplicate entries.

```python
df1.pivot_table(index="Name", columns="Subject", values="Score", aggfunc="mean").rename_axis(columns=None)
```
